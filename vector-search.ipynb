```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma
from langchain.embeddings import HuggingFaceEmbeddings 
```

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
import pandas as pd
```

```python
books = pd.read_csv('books_cleaned.csv')
books.head()
```

```python
books['tagged_description'].to_csv('tagged_description.txt', sep='\n', index=False, header=False)
```

```python
raw_text = TextLoader('tagged_description.txt').load()
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator='\n')
document = text_splitter.split_documents(raw_text)
```

```python
document[0]
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
db_books = Chroma.from_documents(
    document, 
    embedding = embeddings
)
```

```python
query = 'A book to learn about computer science and programming'
docs = db_books.similarity_search(query, k=5)
docs
```

```python
books[books['isbn13'] == int(docs[0].page_content.split()[0].strip())]
```

```python
def retrieve_books(query:str, top_k:int=10) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k=top_k)
    book_list = []

    for i in range(0, len(recs)):
        book_list += [int(recs[i].page_content.split()[0].strip('"'))]

    return books[books['isbn13'].isin(book_list)].head(top_k)
```

```python
retrieve_books('A book to learn about computer science and programming')
```
