```python
category_mapping = {
    'Fiction': 'Fiction',
    'Juvenile Fiction': "children's Fiction",
    'Biography & Autobiography' : 'Nonfiction',
    'History' : 'Nonfiction',
    'Literary Criticism': 'Nonfiction',
    'Philosophy	': 'Nonfiction',
    'Religion': 'Nonfiction',
    'Comics & Graphic Novels': 'Fiction',
    'Drama	': 'Fiction',
    'Juvenile Nonfiction': "children's Nonfiction",
    'Science': 'Nonfiction',
    'Poetry' : 'Fiction',
}

books['simple_categories'] = books['categories'].map(category_mapping)
```

```python
from transformers import pipeline

fiction_categories = ['Fiction', 'Nonfiction']

pipe = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli",
    device=0 
)
```

```python
sequence = books.loc[books["simple_categories"] == "Fiction", "description"].reset_index(drop=True)[0]
```

```python
pipe(sequence, fiction_categories)
```

```python
max_indices = np.argmax(pipe(sequence, fiction_categories)['scores'])
max_label = pipe(sequence, fiction_categories)['labels'][max_indices]
max_label
```

```python
def generate_categories(sequence, categories):
    predictions = pipe(sequence, categories)
    max_indices = np.argmax(predictions['scores'])
    max_label = predictions['labels'][max_indices]
    return max_label
```

```python
actual_categories = []
predicted_categories = []

for i in tqdm(range(0, 300)):
    sequence = books.loc[books['simple_categories'] == "Fiction", "description"].reset_index(drop=True)[i]
    predicted_categories += [generate_categories(sequence, fiction_categories)]
    actual_categories += ['Fiction']
```

```python
for i in tqdm(range(0, 300)):
    sequence = books.loc[books['simple_categories'] == "Nonfiction", "description"].reset_index(drop=True)[i]
    predicted_categories += [generate_categories(sequence, fiction_categories)]
    actual_categories += ['Nonfiction']
```

```python
predictions_df = pd.DataFrame({
    'actual': actual_categories,
    'predicted': predicted_categories
})
```

```python
predictions_df.head(10)
```

```python
predictions_df['corrected_predicted'] = np.where(predictions_df['actual'] == predictions_df['predicted'], 1, 0)
```

```python
predictions_df['corrected_predicted'].sum() / len(predictions_df)
```

```python
isbns = []
predicted_cats = []

missing_cats = books.loc[books['simple_categories'].isna(), ['isbn13', 'description']].reset_index(drop=True)
```

```python

for i in tqdm(range(len(missing_cats))):
    sequence = missing_cats['description'][i]
    predicted_cats += [generate_categories(sequence, fiction_categories)]
    isbns += [missing_cats['isbn13'][i]]
```

```python
missing_predicted_df = pd.DataFrame({
    'isbn13': isbns,
    'predicted_category': predicted_cats
})
```

```python
missing_predicted_df
```

```python
books = pd.merge(books, missing_predicted_df, on='isbn13', how='left')
books['simple_categories'] = np.where(
    books['simple_categories'].isna(),
    books['predicted_category'],
    books['simple_categories']
)
books = books.drop(columns=['predicted_category'])
```
